The strongest legendary and non-legendary pokemon? 

```python
# Importing the requirements
import pandas as pd
df = pd.read_csv('pokemon_data.csv')        
```

```python
df['strength'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed']
[df['strength'].max()]
```

```python
print('Strongest pokemon is: ')
df['Name'].loc[df['strength'] == 780]
```

```python
print('We should stay away from these pokemons: ')
df[df['Type 2'] == 'Poison']
```

```python
legendaries = df[df['Legendary'] == True]
repeats = legendaries['Type 1'].value_counts()
print ('The type that has largest number of legendaries is: ')
print (repeats.index.tolist()[0])
```

```python
nonlegendaries = df[df['Legendary'] == False]
print ('The pokemon types that has no legendaries are: ')
print (nonlegendaries['Type 1'].tolist())
```

```python
print ('The Strongest lengendary pokemon is: ', legendaries['Name'].loc[legendaries['strength'] == legendaries['strength'].max()])
print ('\n')
print ('The Strongest non lengendary pokemon is: ', nonlegendaries['Name'].loc[nonlegendaries['strength'] == nonlegendaries['strength'].max()])
```
